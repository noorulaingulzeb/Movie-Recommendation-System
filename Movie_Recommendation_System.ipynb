```python
# Movie Recommendation System using Collaborative Filtering
# Dataset: MovieLens 100k

import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
print("Downloading dataset...")
```

```python
# MovieLens dataset
ratings_url = "https://files.grouplens.org/datasets/movielens/ml-100k/u.data"
movies_url = "https://files.grouplens.org/datasets/movielens/ml-100k/u.item"

ratings = pd.read_csv(ratings_url, sep="\t", names=["user_id","movie_id","rating","timestamp"])

movies = pd.read_csv(
    movies_url,
    sep="|",
    encoding="latin-1",
    usecols=[0,1],
    names=["movie_id","title"]
)

print("Dataset loaded.")
```

```python
# Merge movie titles with ratings
data = pd.merge(ratings, movies, on="movie_id")

print("Preparing user-movie matrix...")
```

```python
# Create user-movie rating matrix
user_movie_matrix = data.pivot_table(index="user_id", columns="title", values="rating")
```

```python
# Replace NaN with 0
user_movie_matrix = user_movie_matrix.fillna(0)

print("Calculating user similarity...")
```

```python
# Cosine similarity between users
user_similarity = cosine_similarity(user_movie_matrix)

user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_movie_matrix.index,
    columns=user_movie_matrix.index
)

print("Data ready.")
```

```python
# --------------------------------------------------
# Recommendation Function
# --------------------------------------------------

def recommend_movies(user_id, n_recommendations=5):

    similar_users = user_similarity_df[user_id].sort_values(ascending=False)

    similar_users = similar_users.drop(user_id)

    similar_user_id = similar_users.index[0]

    user_movies = user_movie_matrix.loc[user_id]
    similar_user_movies = user_movie_matrix.loc[similar_user_id]

    recommendations = similar_user_movies[
        (similar_user_movies > 4) & (user_movies == 0)
    ]

    return recommendations.sort_values(ascending=False).head(n_recommendations)
```

```python

# --------------------------------------------------
# Generate recommendations
# --------------------------------------------------

user_id = 1

print(f"\nRecommended movies for user {user_id}:\n")

recommendations = recommend_movies(user_id)

for i, movie in enumerate(recommendations.index, 1):
    print(f"{i}. {movie}")
```
